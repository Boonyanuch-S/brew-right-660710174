```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cafethai_prices.csv")
```

```python
df['datetime'] = pd.to_datetime(df['date'] + " " + df['time'], dayfirst=True)
df['date'] = df['datetime'].dt.normalize()

daily_sales = ( df.groupby(['date', 'coffee_name']) .size() .reset_index(name='cups_sold') )
```

```python
def forecast_one(group, periods=60):
  data = (group[['date', 'cups_sold']] .rename(columns={'date': 'ds', 'cups_sold': 'y'}) .sort_values('ds'))
  model = Prophet(daily_seasonality=True, yearly_seasonality=True)
  model.fit(data)
  future = model.make_future_dataframe(periods=periods)
  fcst = model.predict(future).tail(periods)[['ds', 'yhat']]
  fcst = fcst.rename(columns={'ds': 'date', 'yhat': 'predicted_cups'})
  fcst['coffee_name'] = group.name
  return fcst
```

```python
result = (daily_sales .groupby('coffee_name', group_keys=False) .apply(forecast_one, periods=60) .reset_index(drop=True))
```

```python
pivot_result = result.pivot(index='date', columns='coffee_name', values='predicted_cups')
display(pivot_result)
```

```python
for drink, df_drink in result.groupby('coffee_name'):
  plt.figure(figsize=(10, 4))
  plt.plot(df_drink['date'], df_drink['predicted_cups'], label="Predicted Cups")
  plt.title(f"Forecast 60 days - {drink}")
  plt.xlabel("Date")
  plt.ylabel("Cups Sold")
  plt.xticks(rotation=45)
  plt.legend()
  plt.grid(True)
  plt.tight_layout()
  plt.show()
```
